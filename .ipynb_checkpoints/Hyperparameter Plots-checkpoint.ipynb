```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
files = ['12_1', '12_3', '12_6', '24_1', '24_3', '24_6']

for file in files:
    df_results = pd.read_csv('hyperparam_results/{}_res.csv'.format(file))
    print(df_results)
    lstm_results = df_results.pivot(index='hidden dim', columns='learning rate', values = 'MAPE - LSTM')
    lstm_results = 100*lstm_results
    rnn_results = df_results.pivot(index='hidden dim', columns='learning rate', values = 'MAPE - RNN')
    rnn_results = 100*rnn_results
    plt.figure()
    ax = plt.axes()
    sns.heatmap(lstm_results, cbar_kws={'label': 'MAPE'}, ax=ax, vmin=10, annot=True, fmt='.2f')
    ax.set_title('LSTM on {} dataset'.format(file))
    plt.savefig('hyperparam_results/LSTM on {}.png'.format(file))
    plt.figure()
    ax = plt.axes()
    sns.heatmap(rnn_results, cbar_kws={'label': 'MAPE'}, ax=ax, vmin=10, annot=True, fmt='.2f')
    ax.set_title('RNN on {} dataset'.format(file))
    plt.savefig('hyperparam_results/RNN on {}.png'.format(file))
```
